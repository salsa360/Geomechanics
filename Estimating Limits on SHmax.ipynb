## 2. Calculate Maximum horizontal (SHmax) stress values based on Strike-Slip Faulting

```python
# SHmax = maximum horizontal principal stress = ???
# INPUT Variables
# Case 1: Shmin = 6000 psi
# Case 2: Shmin = 6500 psi
# Case 3: Shmin = 7000 psi
# Case 4: Shmin = 8000 psi

Pp = 3000  # Pore pressure (psi)
 = 0.69   # Friction coefficient   
Sh_min = [6000, 6500, 7000, 8000]  # Cases of Shmin stress (psi) - from above

coefficient_ratio = ((**2+1)**(1/2)+)**2
print('Coefficient_ratio =', coefficient_ratio)

def SH_max(Sh_min):
    for i in range(len(Sh_min)):
        Sh_max = coefficient_ratio * (Sh_min[i] - Pp) + Pp
        print ('For Shmin =', Sh_min[i],'psi,', 'SHmax =', Sh_max, 'psi')
        
SH_max(Sh_min)
```
